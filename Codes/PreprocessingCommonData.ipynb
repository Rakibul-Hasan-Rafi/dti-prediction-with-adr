```python
import pandas as pd
```

```python
dti_pd = pd.read_parquet("../Data/Helper_Datasets/scope_onside_common_v2.parquet")
```

```python
# Check null counts per column
print("Null counts before dropping:")
print(dti_pd.isnull().sum())

# Drop nulls
dti_pd = dti_pd.dropna()
```

```python
print(f"\nShape after dropping nulls: {dti_pd.shape}")
print("\nNull counts after dropping:")
print(dti_pd.info())
```

```python
# load meddra data
adr_pd = pd.read_parquet("../Data/Helper_Datasets/final_rxnorm_meddra.parquet")
adr_pd.info()
```

```python
common_rxnorm_ids = dti_pd['rxcui'].unique()

# Create subset of adr_pd
adr_subset = adr_pd[adr_pd['rxnorm_ingredient_id'].isin(common_rxnorm_ids)].copy()

print(f"Original adr_pd shape: {adr_pd.shape}")
print(f"Subset adr_pd shape: {adr_subset.shape}")
print(f"Number of unique rxnorm_ids in common: {len(common_rxnorm_ids)}")
```

```python
adr_subset.to_parquet("../Data/final_rxnorm_meddra_v2.parquet", index=False)
```

```python
valid_rxcui_ids = adr_subset['rxnorm_ingredient_id'].unique()
```

```python
dti_pd_filtered = dti_pd[dti_pd['rxcui'].isin(valid_rxcui_ids)].copy()
```

```python
dti_pd_filtered.to_parquet("../Data/scope_onside_common_v3.parquet", index=False)
```
